This notebook tests the ElevenLabs API for voice synthesis.

- **Best voice quality** among tested providers
- **Low latency** with turbo model
- **Multilingual support** (needed for flexibility)
- **Voice cloning** capability (for custom Yumi voice)

## Environment
Requires `ELEVENLABS_API_KEY` in `.env` file

````python
import os
from IPython.display import Audio, display
from dotenv import load_dotenv
from elevenlabs.client import ElevenLabs
from elevenlabs import VoiceSettings

# Load API key from environment
load_dotenv()
api_key = os.getenv("ELEVENLABS_API_KEY")

if not api_key:
    print("ERROR: ELEVENLABS_API_KEY not found in .env")
    print("Create .env file with: ELEVENLABS_API_KEY=your_key_here")
else:
    # Initialize ElevenLabs client
    client = ElevenLabs(api_key=api_key)
    print("ElevenLabs client initialized!")
    print(f"Using API key: {api_key[:10]}...{api_key[-4:]}")
````

## 1. List Available Voices
Run this cell to see all the voices available in your ElevenLabs account.

````python
# Fetch and display available voices
response = client.voices.get_all()
voices = response.voices

print(f"Found {len(voices)} voices in your ElevenLabs account:\n")
print("-" * 60)

for v in voices:
    print(f'Name: "{v.name}"')
    print(f'ID: {v.voice_id}')
    print("-" * 60)
````

## 2. Test a Voice
Copy a `voice_id` from above and paste it below. Then generate the audio!

````python
# Paste your chosen voice ID here:
chosen_voice_id = "HEhcSPjZtkJJ6iMEEKLW" # Replace with the ID you want to test

demo_text = (
    "W-what are you staring at?! It's not like I dressed up just for you or anything. "
    "I just grabbed whatever was in my closet! "
    "But... since you're here... I guess you can stay. Just don't get any weird ideas, got it?!"
)

print(f"Synthesizing text with voice ID: {chosen_voice_id}...")

try:
    response_chunks = client.text_to_speech.convert(
        voice_id=chosen_voice_id,
        output_format="mp3_22050_32",
        text=demo_text,
        model_id="eleven_multilingual_v2",
        voice_settings=VoiceSettings(
            stability=0.5,
            similarity_boost=0.75,
            style=0.0,
            use_speaker_boost=True,
        ),
    )
    
    # Read all bytes
    audio_data = b"".join([chunk for chunk in response_chunks if chunk])
    
    print("Success! Playing audio...")
    display(Audio(data=audio_data, autoplay=True))
    
except Exception as e:
    print(f"Error generating voice: {e}")
````
