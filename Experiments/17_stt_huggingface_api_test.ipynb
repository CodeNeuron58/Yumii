# HuggingFace Embeddings Test

This notebook tests HuggingFace embeddings for potential RAG implementation.

## What this tests
- HuggingFaceInferenceAPIEmbeddings (API-based)
- HuggingFaceEmbeddings (local models)
- Document embedding and similarity search

## Use case
For adding long-term memory to Yumi through vector storage and retrieval.

## Why not in production
- Local models require significant compute
- API approach adds latency
- Current version uses simple in-memory chat history
- Can be added in future for knowledge base features


## Environment
Requires `HUGGINGFACEHUB_API_TOKEN` in `.env`

````python
from langchain_huggingface.embeddings import HuggingFaceInferenceAPIEmbeddings
import os

# Ensure the API token is available (e.g., set as environment variable before running)
# os.environ["HUGGINGFACEHUB_API_TOKEN"] = "YOUR_API_TOKEN"

embeddings = HuggingFaceInferenceAPIEmbeddings(
    api_key=os.environ["HUGGINGFACEHUB_API_TOKEN"],
    model_name="sentence-transformers/all-MiniLM-l6-v2"
)

text = "This is a test document."
query_result = embeddings.embed_query(text)

print(f"Query result snippet: {query_result[:3]}")
````

````python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings

# Specify the model name (e.g., "sentence-transformers/all-MiniLM-l6-v2")
model_name = "sentence-transformers/all-MiniLM-l6-v2"

# Initialize the embeddings object
hf_embeddings = HuggingFaceEmbeddings(model_name=model_name)

# Embed a document or a query
document_text = "This is an example document for RAG."
query_text = "What is RAG?"

doc_embedding = hf_embeddings.embed_documents([document_text])
query_embedding = hf_embeddings.embed_query(query_text)

print(f"Document embedding length: {len(doc_embedding[0])}")
print(f"Query embedding length: {len(query_embedding)}")
````
