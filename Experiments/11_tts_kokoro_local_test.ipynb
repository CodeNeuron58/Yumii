# Kokoro TTS Test - Local Voice Synthesis

This notebook tests Kokoro, a lightweight local TTS model.

## What is Kokoro?
- Open-source TTS model (~82MB)
- Runs locally (no API calls)
- Fast inference even on CPU
- Good for anime-style voices

## Why not used in production
- Requires GPU for real-time performance
- Voice quality below ElevenLabs
- No voice cloning capability

## Good for
- Offline development/testing
- Reducing API costs during development
- Privacy-sensitive applications

## Dependencies
```bash
pip install kokoro soundfile
```

````python
from kokoro import KPipeline
import soundfile as sf
from IPython.display import Audio
````

````python
# Initialize Kokoro pipeline
# lang_code='a' for American English
# Other options: 'b' (British), 'j' (Japanese), etc.

pipeline = KPipeline(lang_code='a')
print("✓ Kokoro pipeline loaded (82MB model)")
````

````python
# %tb
````

````python
# Synthesize anime-style speech
# Voice 'af_bella' is a female American voice with anime character qualities
# Sample rate: 24kHz (high quality for local TTS)

text = "Hello senpai! I was waiting for you!"

# Generate audio chunks
generator = pipeline(text, voice='af_bella')

# Save to file
for i, (gs, ps, audio) in enumerate(generator):
    sf.write("anime_voice.wav", audio, 24000)

print("Saved to anime_voice.wav")
Audio("anime_voice.wav")
````
