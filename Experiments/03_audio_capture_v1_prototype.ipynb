# Basic Audio Capture Test

This notebook demonstrates basic audio recording using `sounddevice` and visualizes the waveform.

## What this tests
- Microphone access and recording
- Audio file saving (WAV format)
- Basic waveform visualization
- Audio playback in Jupyter

````python
import sounddevice as sd
import numpy as np
from scipy.io.wavfile import write

def record_audio(seconds=5, filename="input.wav", sample_rate=16000):
    """
    Record audio from microphone and save to WAV file.
    
    Args:
        seconds: Recording duration
        filename: Output WAV file path
        sample_rate: Audio sample rate (16kHz is standard for STT)
    
    Returns:
        Path to saved audio file
    """
    print(f"Recording {seconds} seconds of audio...")
    
    # Record audio as float32 array
    audio = sd.rec(
        int(seconds * sample_rate),
        samplerate=sample_rate,
        channels=1,  # Mono audio
        dtype="float32"
    )
    
    # Wait for recording to complete
    sd.wait()
    
    # Save to WAV file (scipy handles the format conversion)
    write(filename, sample_rate, audio)
    
    print(f"Saved to: {filename}")
    return filename

# Record 5 seconds of audio
audio_file = record_audio(5)
````

````python
from IPython.display import Audio

Audio(audio_file)
````

## Audio Analysis

Visualize the waveform and check audio properties to ensure recording worked correctly.

````python
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Load the recorded audio file
rate, data = wavfile.read(audio_file)

# Plot waveform to visualize audio capture
plt.figure(figsize=(12, 3))
plt.plot(data)
plt.title("Audio Waveform")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.show()
````

````python
print(f"Sample rate: {rate} Hz")
print(f"Audio shape: {data.shape}")
print(f"Duration: {len(data) / rate:.2f} seconds")
````
