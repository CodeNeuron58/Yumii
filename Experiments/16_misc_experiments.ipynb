# ElevenLabs Basic Integration Test

Simple test of ElevenLabs API for voice synthesis.

## What this tests
- Basic ElevenLabs API authentication
- Text-to-speech conversion
- Saving audio to file

## Production vs This Test
| Aspect | This Test | Production |
|--------|-----------|------------|
| Model | eleven_turbo_v2_5 | eleven_multilingual_v2 |
| Output | File save | Base64 → WebSocket |
| Voice | Custom | YUMI (custom) |
| Playback | File-based | Real-time streaming |

## Note
API key in this notebook is **REVOKED**. Use `.env` file in production.

```python
import os
import uuid
from dotenv import load_dotenv
from elevenlabs import VoiceSettings
from elevenlabs.client import ElevenLabs

# Load API key from environment (fallback to hardcoded for quick testing)
load_dotenv()

# Note: Hardcoded key is REVOKED - for reference only
ELEVENLABS_API_KEY = os.getenv("ELEVENLABS_API_KEY", "sk_...")

# Initialize ElevenLabs client
elevenlabs = ElevenLabs(api_key=ELEVENLABS_API_KEY)

print("✓ ElevenLabs client initialized")
```

```python
def text_to_speech_file(text: str) -> str:
    """
    Convert text to speech using ElevenLabs and save to MP3 file.
    
    Args:
        text: Text to synthesize
        
    Returns:
        Path to saved MP3 file
    """
    # Call ElevenLabs text-to-speech API
    response = elevenlabs.text_to_speech.convert(
        voice_id="HEhcSPjZtkJJ6iMEEKLW",  # Adam voice (production uses YUMI custom voice)
        output_format="mp3_22050_32",      # 22050Hz, 32kbps MP3
        text=text,
        model_id="eleven_turbo_v2_5",      # Turbo model for low latency
        voice_settings=VoiceSettings(
            stability=0.0,                 # More emotional variation
            similarity_boost=1.0,          # High clarity
            style=0.0,
            use_speaker_boost=True,
            speed=1.0,
        ),
    )

    # Generate unique filename and save audio chunks
    save_file_path = f"{uuid.uuid4()}.mp3"
    with open(save_file_path, "wb") as f:
        for chunk in response:
            if chunk:
                f.write(chunk)

    print(f"Saved: {save_file_path}")
    return save_file_path


# Test with simple text
text_to_speech_file("Hello World")
```
