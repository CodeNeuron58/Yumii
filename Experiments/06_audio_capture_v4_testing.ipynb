````python
from faster_whisper import WhisperModel

model_size = "base"

model = WhisperModel(
    model_size,
    device="cpu",
    compute_type="int8"
)

print("Model loaded")
````

````python
def transcribe_audio(audio_path):

    segments, info = model.transcribe(
        audio_path,
        beam_size=5
    )

    text = ""

    for segment in segments:
        text += segment.text

    return text
````

````python
import time

start = time.time()

text = transcribe_audio("input.wav")

end = time.time()

print("Text:", text)
print("Latency:", end - start, "seconds")
````
