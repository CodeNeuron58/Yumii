````python
import edge_tts
import sounddevice as sd
import numpy as np
import wave
import io
import asyncio

# Edge TTS voice configuration
VOICE = "ja-JP-NanamiNeural"  # Japanese female voice
OUTPUT_FILE = "japanese_voice.mp3"

text = """こんにちは！お元気ですか？私は元気です。"""  # "Hello! How are you? I'm doing well."

async def synthesize_and_play():
    """Synthesize Japanese speech using Edge TTS and play immediately."""
    communicate = edge_tts.Communicate(text, VOICE)
    
    # Save to file first
    await communicate.save(OUTPUT_FILE)
    print(f"Saved to {OUTPUT_FILE}")
    
    # Note: For actual playback, you'd need to decode MP3
    # This is a simplified example
    print("Audio synthesized successfully")

# Run async function
await synthesize_and_play()
````
