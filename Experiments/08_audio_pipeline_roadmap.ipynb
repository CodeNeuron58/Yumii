# Yumi Audio Pipeline Architecture

This notebook outlines the complete audio processing pipeline architecture for Yumi.

## System Flow

```
┌─────────────┐     ┌──────────────────┐     ┌──────────────┐
│   Microphone │────▶│  Speech Detection │────▶│ Audio Capture │
└─────────────┘     │    (WebRTC VAD)   │     └──────────────┘
                    └──────────────────┘            │
                                                    ▼
┌─────────────┐     ┌──────────────────┐     ┌──────────────┐
│   Response  │◄────│  LangGraph Agent  │◄────│    Whisper   │
│    (TTS)    │     │    (LLM/TTS)      │     │     STT      │
└─────────────┘     └──────────────────┘     └──────────────┘
```

## Key Components

1. **Speech Detection**: WebRTC VAD (Voice Activity Detection)
2. **Audio Cleaning**: Silence trimming + normalization
3. **STT**: faster-whisper (local) or Groq Whisper API
4. **Agent**: LangGraph orchestration (Listen → Think → Speak)
5. **TTS**: ElevenLabs voice synthesis

## Future Considerations

- Speaker recognition (see audio-capture-6.ipynb)
- Noise cancellation
- Streaming STT for lower latency

## Pipeline Flow Diagram

Continuous listening cycle:

```
continuous listening
        ↓
speech start detected
        ↓
   capture audio
        ↓
speech end detected
        ↓
   trim silence
        ↓
   speech-to-text
        ↓
    LangGraph
        ↓
       LLM
        ↓
   TTS response
        ↓
return to listening
```
