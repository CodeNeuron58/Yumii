# Groq Orpheus TTS Test

This notebook tests Groq's Orpheus TTS model for voice synthesis.

## Model: canopylabs/orpheus-v1-english

### Pros
- Very low latency (Groq's fast inference)
- Natural speech quality
- Emotional tags support (e.g., [cheerful], [sad])

### Cons
- English only (no multilingual support)
- Limited voice options
- Requires Groq API key

## Why not selected for production
Yumi needs multilingual capability for flexibility. ElevenLabs supports 29 languages.

## Use case
Good for English-only applications where latency is critical.

## Dependencies
```bash
pip install groq
```

## Environment
Requires `GROQ_API_KEY` in `.env` file or environment

````python
import os
from groq import Groq

# Initialize Groq client
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

# Orpheus TTS configuration
model = "canopylabs/orpheus-v1-english"
voice = "troy"  # Voice options: troy, samantha, etc.
response_format = "wav"  # WAV output format

# Test text with emotional tag
text = (
    "Welcome to Orpheus text-to-speech. "
    "[cheerful] This is an example of high-quality English audio "
    "generation with vocal directions support."
)

# Generate speech
response = client.audio.speech.create(
    model=model,
    voice=voice,
    input=text,
    response_format=response_format
)

# Save to file
output_file = "orpheus-english.wav"
response.write_to_file(output_file)
print(f"Saved to {output_file}")
````
