````python
from typecast import Typecast
from typecast.models import TTSRequest, SmartPrompt

# Initialize Typecast client
# Note: API key is revoked - this is for reference only
client = Typecast(api_key="your-Typecast-api-key")

# Synthesize anime character speech
test_text = (
    "Welcome home! I was waiting for you all day! (๑˃̵ᴗ˂̵)و "
    "You look so tired... Did you work super hard again? "
    "You're so amazing, but don't forget to take breaks, okay?~"
)

response = client.text_to_speech(TTSRequest(
    text=test_text,
    model="ssfm-v21",  # Voice synthesis model
    voice_id="tc_6359e7f6467f9e240b68292c",  # Anime character voice
    prompt=SmartPrompt(emotion_type="smart")  # Emotion control
))

# Save audio output
with open('typecast.wav', 'wb') as f:
    f.write(response.audio_data)

print("Saved to typecast.wav")
````
